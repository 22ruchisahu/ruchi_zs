```python
import pandas as pd
```

```python
data=pd.read_csv('/content/athlete_events.csv')
```

```python
data
```

```python
#creating a dataframe
olympics = pd.DataFrame(data)
```

```python
olympics
```

```python
# Counting the distinct values in the 'games' column
total_games = olympics['Games'].nunique()

print("Total distinct games:", total_games)
```

```python
# Retrieving distinct values for 'year', 'season', and 'city' columns, ordered by 'year'
distinct_values = olympics[['Year', 'Season', 'City']].drop_duplicates().sort_values(by='Year')

print("Distinct values for year, season, and city:")
print(distinct_values)
```

```python
data2=pd.read_csv('/content/noc_regions.csv')
```

```python
o_region=pd.DataFrame(data2)
```

```python
all_countries = pd.merge(olympics, o_region, on='NOC')
```

```python
grouped_countries = all_countries.groupby(['Games', 'region']).size().reset_index(name='total_countries')
```

```python
result = grouped_countries.groupby('Games')['total_countries'].count()
```
